```python
# convert ActivityDate to date time format
df['ActivityDate'] = pd.to_datetime(df['ActivityDate'])
df.info()
```

```python
df['TotalMinutes'] = df['VeryActiveMinutes'] + df['FairlyActiveMinutes'] + df['LightlyActiveMinutes'] + df['SedentaryMinutes']
df.columns
```

<h2>Relationship between Calories & Total Steps

```python
# Calories & TotalSteps
fig = px.scatter(df, x = "Calories", y = 'TotalSteps', size = 'TotalMinutes', trendline="ols", 
                    title = "Relationship between Calories & Total Steps")
fig.show()
```

<h3>Observations:</h3>
<p><B>We can see that there is a linear relationship between Total Steps and Calories in day</p>

<h2>Different Minutes states Mean

```python
label = ["Very Active Minutes", "Fairly Active Minutes", 
         "Lightly Active Minutes", "Inactive Minutes"]

counts = df[["VeryActiveMinutes", "FairlyActiveMinutes", 
               "LightlyActiveMinutes", "SedentaryMinutes"]].mean().round(2)


fig = go.Figure(data = [ go.Pie(labels = label, values = counts)])
fig.update_layout(title_text = 'Total Active Minutes')
fig.update_traces(hoverinfo = 'label+percent', textinfo = 'value', textfont_size=25)
fig.show()
```

<H3>Observations:</H3>
<ol><b>
<li>Inactive Minutes is 81.3%</li>
<li>Light Active Minutes is 15.8%</li>
<li>Very Activity Minutes is 1.74%</li>
<li>Fairly Activity Minutes 1.11%</li></ol>
