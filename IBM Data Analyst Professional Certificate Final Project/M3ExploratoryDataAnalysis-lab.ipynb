## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here
df_conv = pd.DataFrame(df['ConvertedComp'])
df_conv.plot(kind = 'box')
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
df['ConvertedComp'].dropna(axis = 0, inplace = True)
```

```python
Q1 = df['ConvertedComp'].quantile(.25)
Q3 = df['ConvertedComp'].quantile(.75) 
```

```python
IQR = Q3 - Q1
IQR
```

Find out the upper and lower bounds.


```python
# your code goes here
upperBound = Q3 + (IQR * 1.5)
lowerBound = Q1 - (IQR * 1.5)
print('Upper Bound is {}\nLower Bound is {}\n'.format(upperBound, lowerBound))
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here
len(df[df['ConvertedComp']< lowerBound]) + len(df[df['ConvertedComp']< upperBound])
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
# your code goes here
dfConv = df['ConvertedComp'].clip(upperBound, lowerBound)
```

```python
dfConv.describe()
```
